```python
qcc=QuantumCircuit(4,3)
qcc.x(3)
qcc.h(3)
qcc.barrier()

for x in range(3):
    qcc.h(x)
qcc.barrier()

qcc=qcc.compose(oraculo_b)
qcc.barrier

for x in range(3):
    qcc.h(x)
qcc.barrier()

qcc.measure([0,1,2],[0,1,2])
qcc.draw()
```

## **EJERCICIO** **6**

 En la construcción más simple del oráculo balanceado, cada qubit de control afecta al qubit objetivo mediante la compuerta CX. Esto asegura que el estado del qubit objetivo se invierta si y solo si todos los qubits de control están en el estado |1>. Esta propiedad es esencial para crear un oráculo que produzca una función balanceada.
Por lo tanto, en el contexto del algoritmo de Deutsch-Jozsa, no es posible prescindir de algún qubit de control en la compuerta CX sin afectar la propiedad necesaria para construir un oráculo balanceado. Cada qubit de control contribuye a determinar si el estado del qubit objetivo debe ser invertido o no.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.circuit.library import XGate,ZGate
from qiskit.visualization import plot_histogram
from qiskit import execute, Aer
```

```python
def oracle_part(qc,qr):
    mcx = XGate().control(3)

    #Los qubits del 0 al 3 son los nodos.
    #Los qubits del 4 al 8 nos permiten reconocer si la arista es correcta
    #(i.e. si se da entre nodos de distinto color).
    qc.cx(qr[0], qr[5])
    qc.cx(qr[3], qr[5])
    qc.barrier()
    qc.cx(qr[0], qr[6])
    qc.cx(qr[4], qr[6])
    qc.barrier()
    qc.cx(qr[1], qr[7])
    qc.cx(qr[3], qr[7])
    qc.barrier()
    qc.cx(qr[1], qr[8])
    qc.cx(qr[4], qr[8])
    qc.barrier()
    qc.cx(qr[2], qr[9])
    qc.cx(qr[3], qr[9])
    qc.barrier()
    qc.cx(qr[2], qr[10])
    qc.cx(qr[4], qr[10])
    qc.barrier()

    #Los qubits del 9 al 11 nos van a permitir contar la cantidad de aristas correctas
    #de tal forma que la suma será expresada en binario con cifras dadas por q_11, q_10 y q_9.
    qc.barrier()
    qc.cx(qr[5],qr[11])
    qc.barrier()

    qc.ccx(qr[6], qr[11], qr[12])
    qc.cx(qr[6],qr[11])
    qc.barrier()

    qc.ccx(qr[7], qr[11], qr[12])
    qc.cx(qr[7],qr[11])
    qc.barrier()

    qc.ccx(qr[8], qr[11], qr[12])
    qc.cx(qr[8],qr[11])
    qc.barrier()
    '''
    qc.append(mcx,[8,11,12,13])
    qc.ccx(qr[7+1], qr[11], qr[12])
    qc.cx(qr[7+1],qr[11])
    qc.barrier()
    '''
    qc.append(mcx,[9,11,12,13])
    qc.ccx(qr[9], qr[11], qr[12])
    qc.cx(qr[9],qr[11])
    qc.barrier()

    qc.append(mcx,[10,11,12,13])
    qc.ccx(qr[10], qr[11], qr[12])
    qc.cx(qr[10],qr[11])
    qc.barrier()

    qc.x(qr[11])
    qc.x(qr[12])
    qc.barrier()

    #El qubit 12 indicará cuando se haya encontrado un estado deseado
    qc.append(mcx,[10+1,11+1,12+1,14])

    return qc
```

```python
qr=QuantumRegister(15)
qc=QuantumCircuit(qr)
oracle_part(qc,qr).draw()
```

```python
def oracle(qc,qr):

    qc=oracle_part(qc,qr)
    qc1=qc.inverse()
    qc.z(14)
    qc.compose(qc1,inplace=True)

    return qc
```

```python
qr=QuantumRegister(15)
qc=QuantumCircuit(qr)
oracle(qc,qr).draw()
```

```python
def grover_diffusion(qc,qr):
    mcz = ZGate().control(4)

    qc.barrier()
    qc.h(range(5))
    qc.x(range(5))
    qc.append(mcz,list(range(0,5)))
    qc.x(range(5))
    qc.h(range(5))
    qc.barrier()
    return qc
```

```python
def grover(qc,qr):
    return grover_diffusion(oracle(qc,qr),qr)
```

```python
qr=QuantumRegister(15)
qc=QuantumCircuit(qr)
grover_diffusion(qc,qr).draw()
```

```python
qr=QuantumRegister(15)
qc=QuantumCircuit(qr)

qra=QuantumRegister(15)
cra=ClassicalRegister(5)
qa=QuantumCircuit(qra,cra)

qa.h(range(5))
qb=grover(qc,qr)
qa.compose(qb.compose(qb),inplace=True)

qa.measure(range(5),cra)

qa.draw()
```

```python
def qft_rotations(circuit, n):
    if n == 0: # Exit function if circuit is empty
        return circuit
    n -= 1 # Indexes start from 0
    circuit.h(n) # Apply the H-gate to the most significant qubit
    for qubit in range(n):
        # For each less significant qubit, we need to do a
        # smaller-angled controlled rotation:
        circuit.cp(pi/2**(n-qubit), qubit, n)
```

```python
qc = QuantumCircuit(4,1)
qft_rotations(qc,4)
qc.draw()
```

```python
def qft_rotations(circuit, n):
    """Performs qft on the first n qubits in circuit (without swaps)"""
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(pi/2**(n-qubit), qubit, n)
    # At the end of our function, we call the same function again on
    # the next qubits (we reduced n by one earlier in the function)
    qft_rotations(circuit, n)

# Let's see how it looks:
qc = QuantumCircuit(4)
qft_rotations(qc,4)
qc.draw()
```

```python
def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    """QFT on the first n qubits in circuit"""
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit

# Let's see how it looks:
qc = QuantumCircuit(4)
qft(qc,4)
qc.draw(output='mpl')
```
